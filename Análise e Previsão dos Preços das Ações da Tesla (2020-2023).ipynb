# O CONTEXTO DO PROBLEMA

### O QUE É O MERCADO DE AÇÕES?

O mercado de ações é um ambiente onde investidores podem comprar e vender participações em empresas públicas. Resumidamente, ele funciona da seguinte forma:

- Empresas públicas: empresas que optam por abrir seu capital vendendo ações ao público em geral. Isso lhes permite captar recursos financeiros para financiar operações, expansão e outros objetivos;

- Investidores: indivíduos e instituições que compram e vendem ações das empresas listadas no mercado de ações. Eles podem incluir investidores individuais, fundos de investimento, bancos, entre outros;

- Bolsas de valores: são os locais físicos ou virtuais onde as transações de compra e venda de ações ocorrem. Algumas das principais bolsas de valores incluem a New York Stock Exchange (NYSE), NASDAQ e Bolsa de Valores de São Paulo (B3);

- Preço das ações: o preço das ações é determinado pela oferta e demanda no mercado. Se muitos investidores querem comprar uma ação, seu preço tende a subir; se muitos querem vender, o preço tende a cair;

- Riscos e retornos: investir em ações pode oferecer potencial para retornos significativos ao longo do tempo, mas também carrega riscos, pois os preços das ações podem ser voláteis e podem ser afetados por uma variedade de fatores, incluindo condições econômicas, desempenho da empresa e eventos geopolíticos;

Em resumo, o mercado de ações é um componente fundamental do sistema financeiro, permitindo que empresas levantem capital e investidores participem do crescimento e sucesso das empresas, com potencial para ganhos financeiros.

# 1.6 - DATA CLEANING E FEATURE ENGINEERING

Nessa etapa, a fim de preparar os dados para a criação de um modelo preditivo, realizei alguns processos envolvendo:
- criação e seleção de variáveis de entrada;
- remoção de variáveis irrelevantes;
- conversão de variáveis para os tipos corretos.

Após uma breve análise das informações obtidas, temos que:
- As variáveis Close e Adj Close possuem valores idênticos. Dessa forma, esta última será removida do dataset;
- Faz-se necessária a criação de duas colunas: a de amplitude (referente a diferença entre os preços máximo e mínimo das ações) e a de valorização (para indicar se a ação valorizou ou desvalorizou em um determinado dia);
- A variável Close será utilizada como a variável dependente (target), isto é, a qual desejamos prever. Diante do fato que ela representa o valor final das ações em um determinado dia, faz todo o sentido que ela seja escolhida para tal objetivo;

```python
# Criação de uma nova coluna no dataset original que representa a diferença entre os preços máximo e mínimo das ações
df['Amplitude'] = df['High'] - df['Low']

# Criação de uma nova coluna no dataset original que representa a valorização (em %) no preço das ações 
df['Valorizacao'] = 100 - (df['Open'] / df['Close']) * 100

# Remoção de uma coluna do dataset original
df.drop("Adj Close", axis = 1, inplace=True)
```

# 3.0 - PREPARAÇÃO DOS DADOS

Nesta etapa, irei preparar os dados para a construção de três modelos de Machine Learning, com foco na predição dos preços das ações da Tesla para o ano de 2024. Tais modelos serão comparados entre si, sendo que aquele que obtiver melhor desempenho será o escolhido.

# 3.1 - IMPORTAÇÃO DAS BIBLIOTECAS

```python
# Importação das bibliotecas necessárias
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
```

# 3.2 - PRÉ-PROCESSAMENTO DOS DADOS

Nesta etapa, irei pré-processar os dados a fim de aumentar a efetividade dos modelos a serem construídos.

- será feita uma padronização dos dados (utilizando o módulo StandardScaler), que transforma cada valor nos dados de forma que a média dos dados padronizados seja zero e o desvio padrão seja igual a um;
- conforme dito anteriormente, a variável Close (correspondente ao valor de fechamento) será a variável dependente;
- as variáveis Close, Date e Valorizacao não farão parte das variáveis preditoras. Enquanto a primeira é o target do modelo, as demais serão irrelevantes para a modelagem do mesmo;
- serão testados três algoritmos de regressão a fim de encontrar o melhor modelo: Random Forest, XGBoost e regressão linear;
- como não há valores vazios/duplicados no dataset, não há a necessidade de tratá-los;
- como apenas variáveis numéricas estão presentes no dataset, não há a necessidade de utilizar o módulo OneHotEncoding (responsável por transformar variáveis categóricas em numéricas);
- as métricas de avaliação utilizadas serão: erro médio absoluto (soma das diferenças absolutas entre os valores previstos e os valores reais, dividido pelo número total de observações) e coeficiente de determinação (também conhecido como R² score);
- quanto menor o valor do erro médio absoluto, mais preciso o modelo é; já quanto maior o valor do coeficiente de determinação, maior é a acurácia.

```python
# Definição das variáveis preditora e dependente, respectivamente
X = df.drop(['Date', 'Close', 'Valorizacao', 'Year', 'Amplitude'], axis=1, errors='ignore')
y = df['Close']

# Criação de uma variável responsável por padronizar os dados
scaler = StandardScaler()

# Criação de duas listas, uma delas vazia
modelos = []
nomes = ["RFR", "XG", "LR"]
```

# 3.3 - SEPARAÇÃO DOS DADOS EM TREINO E TESTE

```python
# Fragmentação dos dados presentes nas variáveis em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Adequando os dados de treino ao padronizador
scaler.fit_transform(X_train)
```

# 3.4 - OBTENDO AS DIMENSÕES DOS DADOS

```python
# Exibição das dimensões dos dados de treino e teste
print(f"Dimensões (X_train): {X_train.shape}")
print(f"Dimensões (y_train): {y_train.shape}")
print(f"Dimensões (X_test): {X_test.shape}")
print(f"Dimensões (y_test): {y_test.shape}")
```

- Os dados estão bem dimensionados.

# 4.0 - MODELAGEM DE DADOS

Nesta etapa, avaliei qual tipo de algoritmo seria o melhor para a construção de um modelo de Machine Learning. No caso apresentado, um modelo baseado em um algoritmo de regressão pareceu ser o mais efetivo na resolução do problema, visto que queremos realizar uma previsão.

Nesse contexto, escolhi três algoritmos de regressão para construir os modelos, sendo eles:

- LinearRegression - também conhecida por regressão linear, trata-se de uma técnica estatística usada para entender a relação entre uma variável dependente (ou resposta) e uma ou mais variáveis independentes (ou preditoras), assumindo uma relação linear entre elas;

- RandomForestRegressor - é uma técnica de aprendizado de máquina que pertence à família de algoritmos de floresta aleatória. Ao contrário do algoritmo anterior, é um método não linear que utiliza múltiplas árvores de decisão para realizar previsões;
  
- XGBRegressor - é um algoritmo de aprendizado de máquina que utiliza o método de Gradient Boosting para realizar tarefas de regressão. O Gradient Boosting é responsável pela produção de um modelo preditivo forte combinando várias árvores de decisão de forma sequencial.

### UTILIZANDO O ALGORITMO RANDOM FOREST REGRESSOR

```python
# Construção de um modelo preditivo utilizando o algoritmo RandomForestRegressor
model_rfr = RandomForestRegressor(n_estimators=500, random_state=42)
model_rfr.fit(X_train, y_train)
predictions_rfr = model_rfr.predict(X_test)
print(f"Erro médio absoluto: {mean_absolute_error(y_test, predictions_rfr)}")
print(f"Precisão: {r2_score(y_test, predictions_rfr)}")
```

### UTILIZANDO O ALGORITMO XGB REGRESSOR

```python
# Construção de um modelo preditivo utilizando o algoritmo xgbRegressor
model_xgb = XGBRegressor(n_estimators=500, learning_rate=0.1)
model_xgb.fit(X_train, y_train)
predictions_xgb = model_xgb.predict(X_test)
print(f"Erro médio absoluto: {mean_absolute_error(y_test, predictions_xgb)}")
print(f"Precisão: {r2_score(y_test, predictions_xgb)}")
```

### UTILIZANDO O ALGORITMO DE REGRESSÃO LINEAR

```python
# Construção de um modelo preditivo utilizando o algoritmo LinearRegression
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
predictions_lr = model_lr.predict(X_test)
print(f"Erro médio absoluto: {mean_absolute_error(y_test, predictions_lr)}")
print(f"Precisão: {r2_score(y_test, predictions_lr)}")
```
