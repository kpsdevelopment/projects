# CONTEXTUALIZAÇÃO - O QUE É A AIDS, COMO ELA PODE SER TRANSMITIDA E QUAIS SEUS PERIGOS?

A AIDS (Síndrome da Imunodeficiência Adquirida) é uma doença grave e potencialmente fatal causada pelo vírus HIV (Vírus da Imunodeficiência Humana). Desde a sua identificação na década de 1980, a AIDS se tornou uma das principais crises de saúde pública no mundo, afetando milhões de pessoas. O HIV ataca e destrói células CD4, que são fundamentais para o sistema imunológico, tornando o corpo vulnerável a uma ampla gama de infecções e certos tipos de câncer que não afetam normalmente pessoas com sistemas imunológicos saudáveis.

A infecção pelo HIV pode permanecer assintomática por muitos anos, mas eventualmente, sem tratamento, a pessoa infectada pode desenvolver AIDS. Os sintomas da AIDS incluem perda de peso significativa, febre recorrente, fadiga extrema, e infecções oportunistas como pneumonia, tuberculose, e certos tipos de câncer como o sarcoma de Kaposi. Esses sintomas são resultado direto do enfraquecimento do sistema imunológico pelo HIV.

A transmissão do HIV ocorre através do contato com fluidos corporais de uma pessoa infectada. As principais vias de transmissão são relações sexuais desprotegidas, compartilhamento de seringas contaminadas, transfusões de sangue contaminado (embora raro em países com sistemas rigorosos de triagem de sangue), e da mãe para o filho durante a gravidez, parto ou amamentação. A prevenção é fundamental e pode ser alcançada através do uso consistente e correto de preservativos, programas de troca de seringas para usuários de drogas injetáveis, e testes regulares de ISTs (Infecções Sexualmente Transmissíveis).

O tratamento da AIDS passou por avanços significativos, especialmente com o desenvolvimento de antirretrovirais (ARVs). Esses medicamentos não curam a infecção pelo HIV, mas podem controlar o vírus, permitindo que as pessoas vivam vidas longas e saudáveis. O tratamento com ARVs reduz a carga viral no corpo a níveis indetectáveis, o que diminui significativamente a probabilidade de transmissão do HIV a outras pessoas. Além disso, a profilaxia pré-exposição (PrEP) é uma medida preventiva eficaz para pessoas em alto risco de infecção pelo HIV.

A resposta global à epidemia de AIDS também inclui educação e conscientização, que são essenciais para a prevenção e redução do estigma associado à doença. Embora o diagnóstico de HIV/AIDS ainda seja um desafio emocional e físico, as campanhas de saúde pública e os avanços médicos trouxeram esperança e melhor qualidade de vida para muitas pessoas afetadas.

A luta contra a AIDS é contínua e requer um esforço colaborativo entre governos, organizações de saúde, e a sociedade civil. A pesquisa continua a buscar uma cura definitiva e uma vacina eficaz, enquanto o foco permanece em fornecer acesso universal a tratamentos que salvam vidas. Através de prevenção, educação, e tratamento eficaz, o impacto da AIDS pode ser significativamente reduzido, aproximando-nos de um futuro onde a AIDS não seja mais uma ameaça à saúde pública.

# PROBLEMAS DE NEGÓCIO E OBJETIVOS DO PROJETO

### PROBLEMAS DE NEGÓCIO

O crescimento da AIDS apresenta diversos desafios que impactam tanto a saúde pública quanto o ambiente de negócios. Um dos principais desafios é a queda na produtividade e o aumento do absenteísmo entre os trabalhadores infectados pelo HIV. Funcionários que sofrem com a doença frequentemente precisam de licenças médicas prolongadas, tratamento contínuo e enfrentam problemas de saúde que comprometem sua capacidade de trabalho. Isso resulta em uma força de trabalho menos eficiente e em custos adicionais para as empresas, que precisam investir em benefícios de saúde e suporte adicional.

Além disso, a escassez de mão de obra qualificada é um problema significativo. A alta mortalidade e a morbidade associadas à AIDS reduzem a disponibilidade de trabalhadores experientes, forçando as empresas a gastar mais em recrutamento e treinamento de novos funcionários. Esse ciclo de contratação e treinamento pode ser particularmente oneroso para pequenas e médias empresas.

A AIDS também afeta o poder de compra das famílias, uma vez que a doença muitas vezes leva à perda de renda, reduzindo a demanda por produtos e serviços. Isso impacta negativamente as receitas das empresas, especialmente aquelas que dependem de mercados locais. No setor agrícola, a redução da força de trabalho devido à doença pode comprometer a produção e a segurança alimentar.

Outro desafio importante é o estigma e a discriminação associados ao HIV/AIDS. Empresas que não implementam políticas inclusivas e de apoio podem enfrentar críticas públicas, deterioração da reputação e problemas legais. Para mitigar esses desafios, é crucial que as empresas adotem políticas abrangentes de apoio aos funcionários, invistam em programas de prevenção e tratamento, e trabalhem em colaboração com governos e organizações de saúde. Somente com um esforço conjunto será possível enfrentar eficazmente os desafios apresentados pelo crescimento da AIDS.

# 1.2 - DICIONÁRIO DE DADOS

Antes de prosseguirmos, é necessário saber o que cada coluna representa:

- time: tempo estimado para insuficiência hepática;

- trt: indicador de tratamento (0 = apenas ZDV; 1 = ZDV + ddl, 2 = ZDV + Zal, 3 = apenas ddl)
  
- age: idade (anos) no início do estudo
  
- wtkg: peso (kg) no início do estudo
  
- hemo: hemofilia (0 = não, 1 = sim)

- homo: atividade homossexual (0 = não, 1 = sim)
  
- drugs: histórico de uso de drogas intravenosas (0 = não, 1 = sim)
  
- karnof: pontuação de Karnofsky (em uma escala de 0 a 100)
  
- oprior: terapia antirretroviral não-ZDV (0 = não, 1 = sim)
  
- z30: uso de ZDV nos 30 dias anteriores aos 175 estipulados (0 = não, 1 = sim)
  
- preanti: dias antes da terapia antirretroviral
  
- race: etnia (0 = branca, 1 = negra)
  
- gender: sexo (0 = feminino, 1 = masculino)
  
- str2: histórico antirretroviral (0 = novo, 1 = experiente)
  
- strat: estratificação do histórico antirretroviral (1 = novo, 2 = > 1 mas menor ou igual a 52 semanas de terapia antirretroviral, 3 = 53 semanas ou mais de terapia antirretroviral)
  
- symptom: indicador sintomático (0 = assintomático, 1 = sintomático)
  
- treat: indicador de tratamento (0 = apenas ZDV, 1 = outros)
  
- offtrt: indicador de falha no tratamento após 5 semanas ou mais (0 = não, 1 = sim)
  
- cd40: CD4 no início do estudo
  
- cd420: CD4 após 5 semanas ou mais
  
- cd80: CD8 no início do estudo
  
- cd820: CD8 após 5 semanas ou mais
  
- infected: está infectado com AIDS (0 = não-infectado, 1 = infectado)


### INFORMAÇÕES EXTRAS

- ZDV/ddl/Zal - medicamentos antirretrovirais usados no tratamento de infecções pelo HIV;
- hemofilia - distúrbio genético e hereditário que afeta a coagulação do sangue;
- pontuação de Karnofsky - forma padrão de medir a capacidade de pacientes com câncer de executar tarefas comuns. Quanto mais alto, melhor;
- terapia antirretroviral - impede a multiplicação do HIV em diferentes etapas de produção no organismo;
- CD4/CD8 - linfócitos responsáveis pela destruição de bactérias, fungos e vírus que invadem o corpo. São os principais alvos do vírus HIV.

# 3.0 - PREPARAÇÃO DOS DADOS

Nesta etapa, irei preparar os dados para a construção de diversos modelos de Machine Learning, com foco em classificar se um determinado paciente possui AIDS, com base em parâmetros que lhes são concedidos. Tais modelos serão comparados entre si, sendo que aquele que obtiver melhor desempenho será o escolhido.

# 3.1 - IMPORTAÇÃO DAS BIBLIOTECAS

```python
# Importação das bibliotecas necessárias
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
```

# 3.2 - PRÉ-PROCESSAMENTO DOS DADOS

Nesta etapa, irei pré-processar os dados a fim de aumentar a efetividade dos modelos a serem construídos.

- será feita uma padronização dos dados (utilizando o módulo StandardScaler), que transforma cada valor nos dados de forma que a média dos dados padronizados seja zero e o desvio padrão seja igual a um;
- o dataset será balanceado através do módulo SMOTE, que utiliza uma técnica que consiste em aumentar a quantidade de registros da classe com menor frequência até que a base de dados possua uma quantidade equilibrada entre as classes da variável alvo. Essa técnica é popularmente conhecida como over-sampling;
- conforme dito anteriormente, a variável infected será a variável dependente; as demais, preditoras;
- serão testados oito algoritmos de classificação distintos: RandomForest, GradientBoosting, AdaBoost, DecisionTree, LogisticRegression (que trabalha com a probabilidade de um evento acontecer), XGBoost, LightGBM e CatBoost;
- a fim de treinar o modelo, os dados (já escalonados e balanceados) serão divididos em variáveis de treino e teste;
- as métricas a serem utilizadas para a avaliação dos modelos são a acurácia e a matriz de confusão. Quanto maior o valor da acurácia, mais preciso o modelo é. Já a matriz de confusão trata-se de uma tabela que mostra as frequências de classificação para cada classe do modelo (verdadeiros, falsos, falsos positivos e falsos negativos);
- será vista posteriormente a importância de cada variável preditora na construção do modelo.

```python
# Definição das variáveis preditora e dependente, respectivamente
X = df.drop("infected", axis=1)
y = df["infected"]

# Criação de uma variável responsável por padronizar os dados
scaler = StandardScaler()

# Criação de uma variável responsável por balancear o dataset
smote = SMOTE()

# Adequando os dados de treino ao padronizador
X_scaled = scaler.fit_transform(X)

# Realizando o balanceamento do dataset
X_sm, y_sm = smote.fit_resample(X_scaled, y)

# Criação de duas listas (uma delas vazia), responsável por armazenar os nomes e resultados de cada modelo
modelos = ['DecisionTreeClassifier', 'LogisticRegression', 'RandomForestClassifier', 'XGBClassifier', 'LGBMClassifier', 'GradientBoostingClassifier', 'CatBoostClassifier', 'AdaBoostClassifier']
scores = []
```

# 3.3 - SEPARAÇÃO DOS DADOS EM TREINO E TESTE

```python
# Fragmentação dos dados presentes nas variáveis em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, shuffle=True, test_size=0.3, random_state=1)
```

# 3.4 - OBTENDO AS DIMENSÕES DOS DADOS

```python
# Exibição das dimensões dos dados de treino e teste
print(f"Dimensões (X_train): {X_train.shape}")
print(f"Dimensões (y_train): {y_train.shape}")
print(f"Dimensões (X_test): {X_test.shape}")
print(f"Dimensões (y_test): {y_test.shape}")
```

- Os dados estão devidamente balanceados e bem dimensionados.

# 4.0 - MODELAGEM DE DADOS

Nesta etapa, avaliei qual tipo de algoritmo seria o melhor para a construção de um modelo de Machine Learning. No caso apresentado, um modelo baseado em um algoritmo de classificação pode ser o mais adequado. Todavia, diante do fato que é quase impossível o modelo acertar todos os casos - são mais de 50000 registros, lembre-se disso - um modelo de regressão logística poderia lidar bem com a situação, visto que ele trabalha mais com a probabilidade de um determinado caso acontecer.

Nesse contexto, escolhi oito algoritmos de classificação para construir os modelos, sendo eles:

- DecisionTreeClassifier - é um algoritmo de aprendizado de máquina supervisionado usado para tarefas de classificação a partir da construção de uma árvore de decisão;
- LogisticRegression - é um modelo estatístico usado para prever a probabilidade de uma variável de resposta binária (duas classes possíveis). Apesar do nome, é um classificador linear, não uma regressão;
- RandomForestClassifier - é um algoritmo de aprendizado de máquina que usa uma coleção (floresta) de árvores de decisão, onde cada árvore é treinada com um subconjunto aleatório dos dados e das características;
- XGBClassifier - é um algoritmo de aprendizado de máquina que utiliza o boosting, uma técnica que combina várias árvores de decisão fracas para formar um modelo robusto;
- LGBMClassifier -  é um algoritmo de aprendizado de máquina baseado em boosting que utiliza uma técnica de construção de árvore de decisão eficiente em termos de memória e tempo;
- GradientBoostingClassifier - é um algoritmo de aprendizado de máquina que cria um modelo preditivo forte ao combinar várias árvores de decisão fracas;
- CatBoostClassifier - é um algoritmo de aprendizado de máquina baseado em boosting, especialmente eficiente com características categóricas;
- AdaBoostClassifier - é um algoritmo de aprendizado de máquina que combina várias árvores de decisão simples (stumps) em sequência.

# 4.1 - CONSTRUÇÃO DOS MODELOS

```python
# Construção de um modelo preditivo utilizando o algoritmo DecisionTreeClassifier
model_dtr = DecisionTreeClassifier()
model_dtr.fit(X_train, y_train)
predictions_dtr = model_dtr.predict(X_test)
print(accuracy_score(y_test, predictions_dtr) * 100)
scores.append(accuracy_score(y_test, predictions_dtr) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo LogisticRegression
model_log = LogisticRegression()
model_log.fit(X_train, y_train)
predictions_log = model_log.predict(X_test)
print(accuracy_score(y_test, predictions_log) * 100)
scores.append(accuracy_score(y_test, predictions_log) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo RandomForestClassifier
model_rfc = RandomForestClassifier()
model_rfc.fit(X_train, y_train)
predictions_rfc = model_rfc.predict(X_test)
print(accuracy_score(y_test, predictions_rfc) * 100)
scores.append(accuracy_score(y_test, predictions_rfc) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo XGBClassifier
model_xgb = XGBClassifier()
model_xgb.fit(X_train, y_train)
predictions_xgb = model_xgb.predict(X_test)
print(accuracy_score(y_test, predictions_xgb) * 100)
scores.append(accuracy_score(y_test, predictions_xgb) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo LGBMClassifier
model_lgb = LGBMClassifier()
model_lgb.fit(X_train, y_train)
predictions_lgb = model_lgb.predict(X_test)
print()
print(accuracy_score(y_test, predictions_lgb) * 100)
scores.append(accuracy_score(y_test, predictions_lgb) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo GradientBoostingClassifier
model_gbc = GradientBoostingClassifier()
model_gbc.fit(X_train, y_train)
predictions_gbc = model_lgb.predict(X_test)
print(accuracy_score(y_test, predictions_lgb) * 100)
scores.append(accuracy_score(y_test, predictions_gbc) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo CatBoostClassifier
model_cat = CatBoostClassifier()
model_cat.fit(X_train, y_train)
predictions_cat = model_cat.predict(X_test)
print()
print(accuracy_score(y_test, predictions_cat) * 100)
scores.append(accuracy_score(y_test, predictions_cat) * 100)
```

```python
# Construção de um modelo preditivo utilizando o algoritmo AdaBoostClassifier
model_ada = AdaBoostClassifier()
model_ada.fit(X_train, y_train)
predictions_ada = model_ada.predict(X_test)
print(accuracy_score(y_test, predictions_ada) * 100)
scores.append(accuracy_score(y_test, predictions_ada) * 100)
```
